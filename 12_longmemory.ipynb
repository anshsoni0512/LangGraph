```python
from langgraph.store.memory import InMemoryStore

# InMemoryStore is an inherited class of BaseStore that stores data in RAM.
```

```python
store  = InMemoryStore()  # create an object of InMemoryStore
```

```python
# create a namespace 
namespace = ("users","u1")   # use tuple to create namespace
```

```python
# NOW CREATE A MEMORY
```

```python
#Adding memories to the store
store.put(namespace,"1",{"data":"user likes pizza"})   # give namespac, key and what type of data to store and value like i want to store dictonary

# this is second memory
store.put(namespace,"2",{"data":"Users prefers grid style navigation"}) 
```

```python
namespace2 = ("user","u2")   # create another namespace for user2

# now add memory u2 folder
store.put(namespace2,"1",{"data":"user2 likes burgers"})
store.put(namespace2,"2",{"data":"user2 prefers list style navigation"})
```

```python
# How to fetch memory from the store
```

```python
store.get(namespace,"1")
```

```python
store.get(namespace2,"2")
```

```python
items = store.search(namespace)
for item in items:
 print(item)
```

```python
items = store.search(namespace)
for item in items:    # items is a list
 print(item.value)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
embedding  = HuggingFaceEmbeddings(
            model_name="BAAI/bge-small-en-v1.5",
        )
```

```python
store = InMemoryStore(index = {"embed":embedding,"dims":1536})

# defining store object 
# now in memory store has a power that it can do semantic search
```

```python
namespace3 = ("person","p1")
```

```python
store.put(namespace3, "1", {"data": "User prefers concise answers over long explanations"})
store.put(namespace3, "2", {"data": "User likes examples in Python"})
store.put(namespace3, "3", {"data": "User usually works late at night"})
store.put(namespace3, "4", {"data": "User prefers dark mode in applications"})
store.put(namespace3, "5", {"data": "User is learning machine learning"})
store.put(namespace3, "6", {"data": "User dislikes overly theoretical explanations"})
store.put(namespace3, "7", {"data": "User prefers step-by-step reasoning"})
store.put(namespace3, "8", {"data": "User is based in India"})
store.put(namespace3, "9", {"data": "User likes real-world analogies"})
store.put(namespace3, "10", {"data": "User prefers bullet points over paragraphs"})
```

```python
items3 = store.search(namespace3,query = "what does user like to study",limit = 1)
for item in items3:
    print(item.value)
```
