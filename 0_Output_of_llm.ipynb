```python
from dotenv import load_dotenv
import os
from langchain_groq import ChatGroq


load_dotenv()
llm = ChatGroq(
    model="llama-3.1-8b-instant",
    groq_api_key=os.getenv('GROQ_API_KEY'),
    temperature = 0,
)
```

```python
from pprint import pprint
pprint( llm.invoke("Tell me about honda cars in 100 chars"),width=100, depth=None, compact=False)
```

```python
llm.invoke("Tell me a joke about computers").content
```

```python
llm.invoke("Tell me a joke about computers").response_metadata
```

```python
llm.invoke("Tell me a joke about computers").response_metadata['model_name']
```

```python
from pydantic import BaseModel, Field
from typing import Literal
class EvaluationExample(BaseModel):
    evaluation:Literal['approved','rejected'] = Field(description="Whether the LinkedIn post is approved or rejected based on the feedback provided.")
    feedback:str = Field(description = "5 wordsFeedback on the LinkedIn post.")

strucuted_model = llm.with_structured_output(EvaluationExample)  # this is our new model that gives only 2 things as output - evaluation and feedback nothing else
```

```python
strucuted_model.invoke("Give feedback on this LinkedIn post: 'I love Agentic ai!'")
```

```python
strucuted_model.invoke("Give feedback on this LinkedIn post: 'I love Agentic ai!'").evaluation
```

```python
# IF YOU WANT TO KEEP RAW DATA IN THE OUTPUT AS WELL THEN include 'include_raw=True' PARAMETER IN THE ABOVE FUNCTION CALL

strucuted_model.invoke("Give feedback on this LinkedIn post: 'I love Agentic ai!'").feedback
```
